```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
import csv
```

```python
driver = webdriver.Firefox()
driver.get('https://www.naukri.com/data-science-jobs-in-mumbai')
```

```python
page_source=driver.page_source
page_source
soup = BeautifulSoup(page_source,'lxml')
soup
```

```python
soup.prettify()
```

```python
soup.body.text
```

```python
mydivs = soup.find_all("div", {"class" : "row" })
```

```python
mydivs
```

```python
divid=[]
for ids in mydivs:
    try:        
        divid.append([ids['data-url'],ids['id'],ids.ul.li['title']])
    except:
        print('Id is not available in Div')
divid
```

```python
df = pd.DataFrame(divid,columns=['URL','ID','Title'])
```

```python
df
```

```python
df.to_csv('Naukari_job_postings.csv')
```

```python
spans = soup.find_all('span',{'class':'exp'})
```

```python
spans[0].text
```

```python
spns=[]
for sp in spans:
    spns.append(sp.text)
spns
```

```python
df1 = pd.DataFrame(spns,columns=['Exp Required'])
```

```python
df1
```

```python
df2 = pd.concat([df,df1],axis=1)
df2
```

```python
spanskill = soup.find_all('span',{'class':'skill'})
```

```python
spanskill[1].text
```

```python
spnskill=[]
for sps in spanskill:
    spnskill.append(sps.text)
spnskill
```

```python
df3 = pd.DataFrame(spnskill,columns=['Key Skills'])
df3
```

```python
df4 = pd.concat([df2,df3],axis=1)
df4
```

```python
spanSal = soup.find_all('span',{'class':'salary'})
```

```python
spanSalr=[]
for spnsl in spanSal:
    spanSalr.append(spnsl.text)
spanSalr
```

```python
df5 = pd.DataFrame(spanSalr,columns=['offered Salary'])
df5
```

```python
df6 = pd.concat([df4,df5],axis=1)
df6
```

```python
df6.to_csv('Naukari_job_posting.csv')
```
